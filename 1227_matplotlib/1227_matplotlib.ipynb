```python
x = np.linspace(0, 10, 10000)#从0到10取1万个值
y = np.sin(x) + 2
plt.plot(x ,y)
```

```python
plt.savefig("sin.pdf", dpi = 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
ax.set(xlabel = "x", ylabel = "sinx")
plt.show()
```

```python
norm = np.random.randn(10000) * 5 * 10
plt.hist(norm, 20)
plt.grid(True)
plt.show()
```

```python
sizes = [10, 20, 15, 5]
labels = ["sing", "dance", "rap", "basketball"]
explodes = [0, 0, 0.1, 0]
fig, ax = plt.subplots()
ax.pie(sizes, labels = labels, explode = explodes, shadow = True, startangle = 90)
ax.axis("equal")
plt.show()
fig.savefig("kunkun.png", dpi = 100)
```

```python
x = np.arange(0, 10, 0.01)
y = np.sin(x) + 2
fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x, y)
ax[0][1].semilogx(x, y)
ax[1][0].semilogy(x, y)
ax[1][1].loglog(x, y)
```
