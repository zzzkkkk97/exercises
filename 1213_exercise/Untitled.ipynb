```python
class A:
    def __init__(self, value):
        self.value = value
    def __enter__(self):
        print("entering A")
        return self
    # 异常类型，值和调用的栈的历史信息
    def __exit__(self, exc_type, exc_value, exc_trace):
        print("exiting A")
with A(10) as a:
    print(a.value)
```

```python
import requests
res = requests.get("https://peps.python.org/api/peps.json")
#https://peps.python.org/api/peps.json
```

```python
print(res.text)
```

```python
# 文本内容保存在文件里
with open("pep.txt", "w") as out:
    #content = out.read()
    #print(content)
    out.write(res.text)
```

```python
import json
```

```python
pep_dict = json.loads(res.text)
```

```python
list(pep_dict.keys())
```

```python
pep_dict["1"]# 字典
```

```python
pep_dict["1"]['authors']
```

```python
# 找到所有已经被接受的pep文档
accepted = []
```

```python
for pep_id in pep_dict:
    if "status" in pep_dict[pep_id] and pep_dict[pep_id]['status'] == "Accepted":
        accepted.append(pep_id)
```

```python
accepted
```

```python
filter(lambda x: "status" in pep_dict[x] and pep_dict[x]['status'] == "Accepted", pep_dict)
```

```python
# 导出并保存json文件
with open("pep.json", "w") as f:
    json.dump(pep_dict, f)
```

```python
pep_dict['0']['authors']
```

```python
header = ["ID", "title", "authors"]
import csv
# newline解决由os造成的在excel文件中读csv时的空行问题
with open("pep.csv", "w", encoding='utf-8', newline = "") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for ID in pep_dict:
        writer.writerow([[ID], pep_dict[ID]['title'], pep_dict[ID]['authors']])
```

```python
# 读出
with open("pep.csv", "r", encoding = 'utf-8', newline = "") as f:
    reader = csv.reader(f)
    content = list(reader)
```
